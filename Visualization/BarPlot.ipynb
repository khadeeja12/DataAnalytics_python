```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('alphabet_stock_data.csv')
data.head(5)
```

```python
new_data=data[(data['Date']>='2020-04-02') & (data['Date']<='2020-09-29')]
plt.figure(figsize=(30, 9))
plt.bar(new_data['Date'],new_data['Volume'],color='orange')
plt.title('Bar Plot between Specific Dates')
plt.xlabel('Date') 
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
