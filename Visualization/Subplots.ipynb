```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    'Date':['2016-03-10','2020-04-02','2020-04-06'],
    'Open':[1122.000000,1098.260010,1098.260010],
    'Close':[1105.619995,1097.880005,1210.280029]
}
df=pd.DataFrame(data)
plt.plot(df['Date'],df['Open'],marker='o',linestyle='-',color='blue')
plt.plot(df['Date'],df['Close'],marker='o',linestyle='-',color='blue')
plt.show()
```
