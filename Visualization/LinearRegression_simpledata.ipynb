```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

```

```python
car_age=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6]).reshape(-1,1)
car_speed=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
```

```python
model=LinearRegression()
model.fit(car_age,car_speed)
```

```python
predictage=model.predict([[10]])
#predictage
print(f"Predicted Speed ={predictage[0]} km/h")
```

```python
r2=model.score(car_age,car_speed)
r2
```

```python
sns.scatterplot(x=car_age.f,y=car_speed,color="blue")
```
